```python
import pandas as pd 
import datetime as dt
import numpy as np
```

```python
data = pd.read_csv("C:/Users/fares/Downloads/archive/online_sales_dataset.csv")
```

```python
data=data[['CustomerID','InvoiceNo','Quantity','InvoiceDate','UnitPrice']]
data['totalPrice'] = data['UnitPrice']*data['Quantity']
```

```python
data=data[(data['Quantity']>0)]
data = data[np.isfinite(data['CustomerID'])]
data['CustomerID']=data['CustomerID'].astype(int)
present=dt.datetime(2024,11,18,00,00)
data['InvoiceDate']=pd.to_datetime(data['InvoiceDate'])
data=data[(data['InvoiceDate']<present)]
```

```python
rfm=data.groupby('CustomerID').agg({'InvoiceDate':lambda date:(present - date.max()).days, 'InvoiceNo':lambda num :len(num),
'totalPrice':lambda price:price.sum()})
rfm.columns=['recency','frequency', 'monetary']
rfm['frequency']=rfm['frequency'].astype(int)
```

```python
rfm['frequency']= rfm['frequency'] + np.random.uniform(0, 1e-6, len(rfm['frequency']))
rfm['rQuartile']=pd.qcut(rfm['recency'],4,['1','2','3','4'])
rfm['fQuartile']=pd.qcut(rfm['frequency'],4,['4','3','2','1'])
rfm['mQuartile']=pd.qcut(rfm['monetary'],4,['4','3','2','1'])
rfm['rfmScore']=rfm['rQuartile'].astype(str)+rfm['fQuartile'].astype(str)+rfm['mQuartile'].astype(str)

```

```python
print(rfm[rfm['rfmScore']=='111'].sort_values('monetary',ascending=False).head())
```
